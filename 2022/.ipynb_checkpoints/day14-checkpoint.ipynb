Part 01

```python
def draw_scenario(paths, correct_misread=False):
    min_x, max_x, max_y = paths[0][0][0], paths[0][0][0], paths[0][0][1]
    for path in paths:
        for point in path:
            min_x, max_x, max_y = min(min_x, point[0]), max(max_x, point[0]), max(max_y, point[1])

    if correct_misread:
        max_y+=2
        min_x=min(min_x, 500-max_y-1)
        max_x=max(max_x, 500+max_y+1)
    
    
    matrix = [['.']*(max_x-min_x+1) for _ in range(max_y+1)]
    
    for path in paths:
        for path_idx in range(len(path)-1):
            point1 = path[path_idx][::-1]
            point2 = path[path_idx+1][::-1]
            if point1[0] == point2[0]:
                for j in range(point2[1] if point1[1] > point2[1] else point1[1], point1[1]+1 if point1[1] > point2[1] else point2[1]+1, +1):
                    matrix[point1[0]][j-min_x] = '#'
            else:
                for i in range(point2[0] if point1[0] > point2[0] else point1[0], point1[0]+1 if point1[0] > point2[0] else point2[0]+1, +1):
                    matrix[i][point1[1]-min_x] = '#'

    if correct_misread:
        matrix[-1] = ['#']*len(matrix[0])

    return matrix, [0, 500-min_x]
```

```python
def units_of_sand(matrix, sand_unit_source):

    def swap(matrix, orig, dest):
        matrix[orig[0]][orig[1]], matrix[dest[0]][dest[1]] = matrix[dest[0]][dest[1]], matrix[orig[0]][orig[1]]

    def fall_forever(matrix, next_cell):
        if next_cell[0] == len(matrix) or next_cell[1] == -1 or next_cell[1] == len(matrix[0]):
            return True
        else:
            return False

    iteration = 0
    while(True):
        sand_unit_cell = [sand_unit_source[0], sand_unit_source[1]]
        if matrix[sand_unit_cell[0]][sand_unit_cell[1]] != '.':
            return iteration
        else:
            matrix[sand_unit_cell[0]][sand_unit_cell[1]] = '+'

        while(matrix[sand_unit_cell[0]][sand_unit_cell[1]] != 'o'):

            next_possible_cells = [[sand_unit_cell[0]+1, sand_unit_cell[1]],\
                                   [sand_unit_cell[0]+1, sand_unit_cell[1]-1],\
                                   [sand_unit_cell[0]+1, sand_unit_cell[1]+1]]

            sand_unit_flows = False
            
            for next_possible_cell in next_possible_cells:
                if fall_forever(matrix, next_possible_cell):
                    return iteration
                if matrix[next_possible_cell[0]][next_possible_cell[1]] == '.':
                    swap(matrix, sand_unit_cell, next_possible_cell)
                    sand_unit_cell = next_possible_cell
                    sand_unit_flows = True
                    break
                    
            if not sand_unit_flows:
                matrix[sand_unit_cell[0]][sand_unit_cell[1]] = 'o'
                
        iteration+=1
```

```python
with open('input14.csv') as csvfile:
    paths = [[[int(y) for y in x.split(',')] for x in path.strip().split(' -> ')] for path in csvfile.readlines()]

matrix, sand_unit_source = draw_scenario(paths)

answer = units_of_sand(matrix, sand_unit_source)
print(answer)

for row in matrix:
    print(''.join(row))
```

```python
with open('input14.csv') as csvfile:
    paths = [[[int(y) for y in x.split(',')] for x in path.strip().split(' -> ')] for path in csvfile.readlines()]

matrix, sand_unit_source = draw_scenario(paths, correct_misread=True)

answer = units_of_sand(matrix, sand_unit_source)
print(answer)

for row in matrix:
    print(''.join(row))
```
